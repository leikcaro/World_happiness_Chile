__Chile en World Happiness 2016-2020__
 
Este análisis no tiene otro objetivo que aportar, con una opinión personal, de como ha evolucionado Chile en torno a la felicidad de su población y qué factores son los que han sido determinantes para retroceder o avanzar en el ranking, lo que se refleja en qué hace más o menos felices a sus habitantes 
Estos son los hallazgos de una encuesta de Ipsos en 27 países realizada del 24 de julio al 7 de agosto de 2020 entre 19,516 adultos de 18 a 74 años en los Estados Unidos, Canadá, Malasia, Sudáfrica y Turquía y de 16 a 74 en otros 22 países, a través de Plataforma de encuestas en línea Global Advisor de Ipsos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

wh2020=pd.read_csv("wh_2020.csv")
wh2019=pd.read_csv("wh_2019.csv")
wh2018=pd.read_csv("wh_2018.csv")
```

```python
#creamos una columna con el año correspondiente a cada DF para no confundir los valores y su ranking

wh2018["ranking_in_year"]=wh2018.index+1
wh2018["Year"]=2018
wh2019["ranking_in_year"]=wh2019.index+1
wh2019["Year"]=2019
wh2020["ranking_in_year"]=wh2020.index+1
wh2020["Year"]=2020
wh2020
```

```python
#revisaremos las columnas y sus nombres, para descartar las columans que no usaremos y dejarlos homogeneos
wh2020=wh2020.drop(['upperwhisker', 'lowerwhisker', 'Logged GDP per capita', 'Explained by: Social support',
       'Healthy life expectancy',
       'Freedom to make life choices',
       'Generosity', 'Perceptions of corruption',
       'Dystopia + residual'], axis=1)
wh2020.columns
```

```python
wh2019.head()
cols_2019={"Country or region":"Country name", "Score":"Ladder score", "GDP per capita":"GDP"}
wh2019=wh2019.rename(mapper=cols_2019, axis=1)
wh2019
wh2019.drop("Overall rank", axis=1)
```

```python
cols_2020={'Explained by: Log GDP per capita':"GDP", "Explained by: Healthy life expectancy":"Healthy life expectancy", "Explained by: Freedom to make life choices":"Freedom to make life choices",'Explained by: Generosity':"Generosity", 'Explained by: Perceptions of corruption':"Perceptions of corruption",'Explained by: Social support':"Social support"}
wh2020=wh2020.rename(mapper=cols_2020, axis=1)
wh2020=wh2020.drop(["Regional indicator", "Standard error of ladder score", "Ladder score in Dystopia"], axis=1)
wh2020
```

```python
cols_2018={"Country or region":"Country name", "Score":"Ladder score", "GDP per capita":"GDP"}
wh2018=wh2018.rename(mapper=cols_2018, axis=1)
wh2018=wh2018.drop("Overall rank", axis=1)
wh2018
```

```python
combined_2018_2019=wh2018.merge(wh2019, how="outer")
combined=combined_2018_2019.merge(wh2020, how="outer")

combined.head()
```

```python
Chile= combined[combined["Country name"]=="Chile"]
Chile
```

```python
combined.plot()
```

Como primera impresion, destaca el __drástico descenso en el ranking mundial de Chile, pasando de el puesto 24 en el 2018, al 25 en el 2019 al 33 el año 2020__ 
Comparemos este resultado con algunos de los paises de la región

```python
sudamerica=["Chile", "Argentina", "Peru","Uruguay",]
combined_latam=combined[combined["Country name"].isin(sudamerica)]
combined_latam.set_index("Country name")
```

```python
combined_latam["Year"]=combined_latam["Year"].astype("str")+" "+combined_latam["Country name"]
```

```python
combined_latam.plot(kind="barh", x="Year", y="Social support")
```

```python
#el impacto que tuvo el coronavirus en la percepcion de soporte social desde sus gobiernos ha sido muy grande, pero más en paises como Argentina y Uruguay
combined_latam.plot(kind="barh", x="Year", y="Ladder score")
```

```python
#se observa que Perú y Uruguay se han mantenido estables en su ranking, Argentina ha reducido bastante su posicion, y Chile levemente a la baja
combined_latam.plot(kind="barh", x="Year", y="Healthy life expectancy")
```

```python
#Hubo una reduccion moderada pese a lo que se podria pensar, de la esperanza de vida saludable 
combined_latam.plot(kind="barh", x="Year", y="Freedom to make life choices")
```

```python
#sorpresivamente la sensación de libertada para tomar desiciones de vida ha subido en chile y argentina.
combined_latam.plot(kind="barh", x="Year", y="Generosity")
```

```python
#la generosidad en Chile se ha visto a la baja año a año moderadamente
combined_latam.plot(kind="barh", x="Year", y="Perceptions of corruption")
```
